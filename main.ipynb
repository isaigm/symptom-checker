```python
import xgboost as xgb
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np

df = pd.read_csv("dataset.csv")

df.head()
```

##### Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score


label_encoder = LabelEncoder()
y =  label_encoder.fit_transform(df["prognosis"])

x = df.drop("prognosis", axis='columns')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

##### Model training and evaluation

```python
model = xgb.XGBClassifier(n_estimators=1, max_depth=1, objective='multi:softmax')
scores = cross_val_score(model, x, y, cv=20)
model.fit(x_train, y_train)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

##### Inference

```python
test_input = np.zeros(131)
test_input[0] = 1
test_input[1] = 1
test_input[10] = 1
test_input[11] = 1
test_input[13] = 1

probas = model.predict_proba([test_input])[0] * 100
idx = (-probas).argsort()[:3]

result = dict(zip(label_encoder.classes_[idx], probas[idx]))
print(result)
```
